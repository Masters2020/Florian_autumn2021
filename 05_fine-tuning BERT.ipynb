## **Model Tuning - Masked Language Modelling**

### **Preparation**

```python
import datasets
from datasets import load_dataset
from datasets import Dataset
from transformers import AutoTokenizer, BertTokenizer
from transformers import AutoModelForMaskedLM
from transformers import Trainer, TrainingArguments
from transformers import DataCollatorForLanguageModeling
import math
from datasets import ClassLabel
import random
from IPython.display import display, HTML
```

```python
from transformers import set_seed
set_seed(42)
```

```python
#this will define which model and tokenizer are used, change if necessary. Consult huggingface documentation to see other options
model_checkpoint = 'bert-base-uncased'
```

### **Data Import and EDA**

```python
train_red = '/content/drive/MyDrive/data/you_cons_train.csv' #change to location train data
val_red = '/content/drive/MyDrive/data/you_cons_val.csv' #change to location val data
```

```python
datasets = load_dataset('csv', data_files = {'train': train_red, 'val': val_red})
```

```python
lent = len(datasets['train'])
lenv = len(datasets['val'])
print('train length:', lent, 'val length:', lenv, 'overall:', lent+lenv)
```

```python
datasets['train'][10]
```

### **Setting tokenizer and input data**

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, use_fast=True)
```

```python
#add new entities to the tokenizer so they are not split and can be used later on
vocab = '/content/drive/MyDrive/data/you_NER_vocab_10.txt' #change to location of vocab of entities
with open(vocab, 'r', encoding = 'utf-8') as file1:
  new_tokens = file1.readlines()
clean_tokens = []
for i in new_tokens:
  j = i.replace('\n', '')
  clean_tokens.append(j)
```

```python
print(len(clean_tokens))
```

```python
tokenizer.add_tokens(clean_tokens)
```

```python
def tokenize_function(examples):
    return tokenizer(examples['text'])
```

```python
tokenized_datasets = datasets.map(tokenize_function, batched=True, remove_columns=["text"])
```

```python
lentt = len(tokenized_datasets['train'])
lenvt = len(tokenized_datasets['val'])
print('train length:', lentt, 'val length:', lenvt, 'overall:', lentt+lenvt)
```

```python
# block_size = tokenizer.model_max_length
block_size = 128
```

```python
def group_texts(examples):
    # Concatenate all texts.
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    # We drop the small remainder, we could add padding if the model supported it instead of this drop, you can
        # customize this part to your needs.
    total_length = (total_length // block_size) * block_size
    # Split by chunks of max_len.
    result = {
        k: [t[i : i + block_size] for i in range(0, total_length, block_size)]
        for k, t in concatenated_examples.items()
    }
    result["labels"] = result["input_ids"].copy()
    return result
```

```python
lm_datasets = tokenized_datasets.map(
    group_texts,
    batched=True,
    batch_size=1000,
)
```

```python
tokenizer.decode(lm_datasets["train"][1]["input_ids"])
```

```python
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm_probability=0.15)
```

### **Setting model and depository**

```python
model = AutoModelForMaskedLM.from_pretrained(model_checkpoint)
```

```python
model.resize_token_embeddings(len(tokenizer)) 
```

```python
model_name = 'ENTER MODEL NAME HERE'
training_args = TrainingArguments(
    f"{model_name}",
    evaluation_strategy = "epoch",
    learning_rate=2e-5,
    weight_decay=0.01,
    push_to_hub = True
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=lm_datasets["train"],
    eval_dataset=lm_datasets["val"],
    data_collator=data_collator,
)
```

### **Tuning and evaluating model**

```python
trainer.train()
```

```python
eval_results = trainer.evaluate()
print(f"Perplexity: {math.exp(eval_results['eval_loss']):.2f}")
```

### **Upload fine-tuned model to repo**

```python
trainer.push_to_hub()
```
